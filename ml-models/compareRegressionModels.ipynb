```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data upload
data = pd.DataFrame({
    'Location': ['London UK', 'New York US', 'Odesa Ukraine'],
    'Year': [2023, 2023, 2022],
    'Month': [2, 2, 10],
    'Day': [7, 3, 16],
    'Hour': ['8:30', '12:00', '6:30'],
    'decisiontree': [43, 500, 185],
    'randomforest': [78, 480, 199],
    'xgboost': [26, 91, 10],
    'dnn': [258, 78, 52],
    'lstm': [10, 17, 12],
    'mlp': [17, 47, 26]
})

# Select only columns with model results
model_columns = ['decisiontree', 'randomforest', 'xgboost', 'dnn', 'lstm', 'mlp']
df_models = data[model_columns]
```

```python
# 1. Boxplot - distribution of model predictions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_models)
plt.title("Distribution of forecasts from different models")
plt.ylabel("Predicted AQI value")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Line graph of forecast changes over time
plt.figure(figsize=(10, 6))
for model in model_columns:
    plt.plot(data['Day'], data[model], marker='o', label=model)

plt.title("AQI forecast by models depending on date")
plt.xlabel("Day")
plt.ylabel("Prediction AQI")
plt.legend()
plt.show()
```

```python
# 3. Histograms of the distribution of model predictions
plt.figure(figsize=(12, 6))
for model in model_columns:
    sns.kdeplot(data[model], label=model, fill=True)

plt.title("Distribution of model prediction values")
plt.xlabel("Predicted AQI value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# 4. Correlation matrix of models
plt.figure(figsize=(8, 6))
sns.heatmap(df_models.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation matrix of model predictions")
plt.show()
```
